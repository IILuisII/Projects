```python
import matplotlib
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import psycopg2
```

```python
engine = create_engine("postgres://readonly:12345@35.225.151.235:5432/postgres")
inspector = inspect(engine)
inspector.get_table_names()
```

```python
conn = engine.connect()
```

```python
query = """
            SELECT
                *
            FROM
                salaries
        """
```

```python
df = pd.read_sql(query, conn)
df
```

```python
query2 = """
            SELECT
                *
            FROM
                employees
        """
```

```python
df2 = pd.read_sql(query2, conn)
df2
```

```python
query3 = """
            SELECT
                *
            FROM
                titles
        """
```

```python
df3 = pd.read_sql(query3, conn)
df3
```

```python
# Merge Columns / Combine Data
combined_data = pd.merge(df, df2, on="emp_no", how="inner")
combined_data
```

```python
combined_data2 = pd.merge(combined_data, df3, left_on="emp_title_id", right_on="title_id", how="inner")
combined_data2.head()
```

```python
# Groupby Title
grouped_df = combined_data2.groupby("title").mean()
grouped_df
```

```python
newdf = grouped_df.drop(columns =['salary_id', 'emp_no']).reset_index()
newdf
```

```python
#bins = [40000, 40001-50000, 50001-60000, 60001-70000, 70001-80000, 80000]
plt.hist(df.salary, bins=10, color="r")
plt.grid(False)
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary Frequency")
plt.show()
```

```python
plt.bar(newdf["title"], newdf["salary"], color='r', align="center")
plt.xticks(rotation=45)
plt.grid(False)
plt.xlabel("Title")
plt.ylabel("Salary")
plt.title("Average Salary By Title")
plt.show()
```
